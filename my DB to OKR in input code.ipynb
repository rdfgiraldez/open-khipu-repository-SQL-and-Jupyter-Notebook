```python
import pandas as pd
import sqlalchemy as s
import pandas as pd
engine = s.create_engine('sqlite:///AS1 - AS9.db')
```

```python
def lookupL(l):
    df = pd.read_sql_query('SELECT * from '+ str(l),engine)
    return df
def to_string(x):
    if type(x) is str:
        return '"'+x+'"'
    elif x is None:
        return '""'
    else:
        if str(x)[0] == 0 and len(str(x))>5:
            return '"'+x+'"'
        return str(x)
```

```python
df = lookupL('ascher_cord_color')

print('''INSERT INTO ascher_cord_color (color_id,khipu_id, cord_id, pcord_flag,CREATED_BY,
        CREATED_ON,CHANGED_BY,CHANGED_ON, color_range, color_cd_1,operator_1
        , color_cd_2,operator_2, color_cd_3,operator_3, color_cd_4,operator_4
        , color_cd_5, range_beg,range_end, full_color) values ''')
for index, row in df.iterrows():
    col = ['color_id','KHIPU_ID','CORD_ID','PCORD_FLAG','CREATED_BY','CREATED_ON','CHANGED_BY','CHANGED_ON',\
           'COLOR_RANGE','COLOR_CD_1','OPERATOR_1','COLOR_CD_2','OPERATOR_2','COLOR_CD_3','OPERATOR_3','COLOR_CD_4','OPERATOR_4','COLOR_CD_5',\
            'RANGE_BEG','RANGE_END','FULL_COLOR']
    vals = '('
    for c in col:
        vals = vals+to_string(row[c])+', '
    vals = vals[:-2]+ '),'
    print(vals)
    
```
